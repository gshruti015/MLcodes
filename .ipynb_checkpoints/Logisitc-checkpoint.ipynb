```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import uniform
import math
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import random
import sys
pd.options.display.max_seq_items=5000
pd.options.display.max_rows=20
pd.options.display.max_columns=150
np.set_printoptions(threshold=sys.maxsize)
```

### Drawing a S shaped curve using the sigmoid function at diff values of theta

```python
random1=[]
random2=[]
random3=[]
xlist = []
theta=[10, 1,0.1]
for i in range(100):
 x = uniform(-5,5)
 xlist.append(x)
 logreg1 = 1/(1+math.exp(-(theta[0]*x)))
 logreg2 = 1/(1+math.exp(-(theta[1]*x)))
 logreg3 = 1/(1+math.exp(-(theta[2]*x)))
 random1.append(logreg1)
 random2.append(logreg2)
 random3.append(logreg3)
plt.scatter(xlist, random1, marker='*',s=40, c='orange',alpha=0.5,label=r'$\theta = %3.1f$'%(theta[0]))
plt.scatter(xlist, random2, c='magenta',alpha=0.3,label=r'$\theta = %3.1f$'%(theta[1]))
plt.scatter(xlist, random3, c='navy',marker='d', alpha=0.3,label=r'$\theta = %3.1f$'%(theta[2]))
plt.axhline(y=0.5, label='P=0.5')
plt.ylabel(r'$P=\frac{1}{1+e^{-\theta \, x}}$', fontsize=19)
plt.xlabel(r'$x$',fontsize=18)
plt.legend(fontsize=16)
plt.show()
```

## Plot showing how not having negative x values in the same function does not give s shaped curve, or gives a near straigt line

```python
x = [-0.5,-0.37,0.35,0.20,0.10,0.1]
# x = np.random.normal(30, 2, 20)
random1=[]
for a in x:
    xlist.append(a)
    logreg1 = 1/(1+math.exp(-(theta[0]*a)))
    random1.append(logreg1)
plt.scatter(x, random1, marker='*',s=40, c='orange',alpha=0.5,label=r'$\theta = %3.1f$'%(10))
```

## Applying Logisitc through scikit learn

```python
length = 20
Salary = [random.choice([-1300,-1600,-1800,-1900,2300,2500,2700,3300,4000]) for x in range(length)]
DataFrame=pd.DataFrame({'Salary':Salary})

DataFrame['output'] = np.where((DataFrame.Salary > 2000),1,0)

target = DataFrame.loc[:,'output']
data = DataFrame.loc[:,['Salary']]

clf = LogisticRegression(random_state=0)
clf = clf.fit(data, target)
```

## predicting single value of salary

```python
clf.predict([[2300]])
```

## predicting on data

```python
clf.predict(data)
```

## Getting optimal parameter values 

```python
clf.coef_[0][0]
```

```python
clf.intercept_[0]
```

```python
clf.predict_proba(data)
```

## plotting the above data reg line, gives kind of an S curve

```python
random1=[]
for a in data['Salary']:
    xlist.append(a)
    logreg1 = 1/(1+math.exp(-(clf.intercept_[0]+clf.coef_[0][0]*a)))
    random1.append(logreg1)
plt.scatter(data['Salary'], random1, marker='*',s=40, c='orange',alpha=0.5,label=r'$\theta = %3.1f$'%(theta[0]))
```

## these results will match with the results of clf.predict_proba(data) for class 1's probabilities
